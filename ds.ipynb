```python
order_df = pd.read_csv(ofile, sep='|', header=None)
order_df.columns =['session', 'orderid', 'side', 'action', 'order_time', 'price', 'fquant', 'dquant', 'market', 'ioc', 'stoploss', 'triggerprice']
order_df
```

```python
jiffy = 65536
ist = 330*60
order_df['time'] = order_df['order_time'].apply(lambda x: datetime.fromtimestamp(x/jiffy-ist))#.strftime('%Y-%m-%d %H:%M:%S'))
order_df['min'] = order_df['order_time'].apply(lambda x: datetime.fromtimestamp(x/jiffy-ist).strftime('%H:%M'))
order_df
```

```python
(order_df.groupby('min').count())['orderid'].to_csv('orders_per_minute.csv')
```

```python
ohlc = []
non_market = order_df.groupby('market').get_group('N')
for t in order_df['min'].unique():
    try:
        ohc = non_market.groupby('min').get_group(t)['price']
        # l = order_df.groupby('market').get_group('N').groupby('min').get_group(t)['price']
        ohlc.append(((t, ohc.iloc[0] , ohc.max(), ohc.min(), ohc.iloc[-1])))
    except KeyError:
        ohlc.append(((t, np.nan, np.nan, np.nan, np.nan)))

ohlc

```
