```python
import multiprocessing

def orders_per_minute(df, tm):
    ord = df['price'][df['time'].str.contains(tm)==True].shape[0]
    return (tm, ord)

results =  []
def collect_results(result):
    results.extend(result.values.tolist())

pool = multiprocessing.Pool(processes=multiprocessing.cpu_count())
for tm in fin:
    pool.apply_async(orders_per_minute, args=(order_df, tm), callback=collect_results)
pool.close()
pool.join()  

results
```
